```python
import numpy as np
from numpy import mean
from numpy import std
import pandas as pd
import matplotlib.pyplot as plt




```

```python
#=========================================================================
# read data
#=========================================================================
data_csv = pd.read_csv('../data/curated/2017_2022_house_data_final.csv')
```

```python
data_csv.describe()
```

```python
#=========================================================================
# check null
#=========================================================================
data_csv.isnull().sum()
```

```python
#=========================================================================
# remove unrealistic data and check the data distribution
#=========================================================================
data_csv = data_csv[(data_csv.week_rent > 50) \
                   &(data_csv.week_rent < 100000)]
data_csv.describe()
```

```python

data_csv
```

```python
import numpy as np
from numpy import mean
from numpy import std
import matplotlib.pyplot as plt
#=========================================================================
# Differentiation of data by years.
#=========================================================================
history = data_csv
data_22 = history[history['years']==2022]
data_21 = history[history['years']==2021]
data_20 = history[history['years']==2020]
data_19 = history[history['years']==2019]
data_18 = history[history['years']==2018]
data_17 = history[history['years']==2017]
data_year = [data_21,data_20,data_19,data_18,data_17]

year = [2022,2021,2020,2019,2018,2017]
i=0

#=========================================================================
# Record the outliers by bedroom number and years.
#=========================================================================
recorder = []
for data in data_year:
    for bed_num in list(data['bed'].unique()):
        if pd.isna(bed_num):
            continue        
        da = data[data['bed']==bed_num]
        data_mean, data_std = mean(da['week_rent']), std(da['week_rent'])
        # identify outliers
        cut_off = data_std * 3
        lower, upper = data_mean - cut_off, data_mean + cut_off
        outliers = da[(da['week_rent'] > upper) | (da['week_rent'] < lower)]
        recorder.append(outliers)
        print(f'year {year[i]} have {len(outliers)} outliers(total number: {len(da)}) in the apartment with {bed_num}')
    i+=1
```

```python
#=========================================================================
# Remove the outliers.
#=========================================================================
z = 0
for j in range(0, len(data_csv)-z):
    for k in range(0, len(recorder)):
        for a in range(0, len(recorder[k])):
            if data_csv.iloc[j]['Unnamed: 0'] == recorder[k].iloc[a]['Unnamed: 0']:
                z = z+1
                data_csv = data_csv.drop(data_csv.iloc[j]['Unnamed: 0'], axis=0)
```
