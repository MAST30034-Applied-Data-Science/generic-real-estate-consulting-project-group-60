```python
import pandas as pd
```

```python
data = pd.read_csv("../data/raw/SA2.csv")
data
```

```python
#only for victoria#
data = data.loc[data['S/T code'] == 2]
data
```

```python
#SA2 name and code in VIC#
SA2 = data[['SA2 code', 'SA2 name']]
SA2
```

```python
income = pd.read_csv("../data/raw/2021Census_G32_VIC_SA2.csv")

```

```python
income = income[['SA2_CODE_2021','Tot_Tot']]
income
```

```python
income_merge = pd.merge(SA2, income ,how = 'inner', left_on='SA2 code',right_on ='SA2_CODE_2021')
income_merge = income_merge.drop(['SA2_CODE_2021'], axis=1)
income_merge = income_merge.rename(columns={'Tot_Tot':'Total family income weekly'})
income_merge
```

```python
population = pd.read_csv('../data/raw/population_2021.csv')
population.head()
```

```python
population_vic= population.loc[population['S/T code'] == 2]
```

```python
population_vic = population_vic[['SA2 code','no..16','no..17','no..18','no..19','no..20']]
population_vic = population_vic.rename(columns={'no..20':'2021 Population',
                                                  'no..19':'2020 Population',
                                                  'no..18':'2019 Population',
                                                  'no..17':'2018 Population',
                                                  'no..16':'2017 Population'})
```

```python
inc_pop_merge = pd.merge(income_merge, population_vic ,how = 'inner', left_on='SA2 code',right_on ='SA2 code')
#inc_pop_merge = inc_pop_merge.drop(['2021 Population_y'], axis=1)
#inc_pop_merge = inc_pop_merge.rename(columns={'2021 Population_x':'2021 Population'})
inc_pop_merge

```
