# Visualization of API Explanation

```python
# Import the function needed into the notebook
import folium
import pandas as pd
import openrouteservice
from openrouteservice import convert
```

# Using one route as example

```python
# Read the School location files
# Read the Historical Property data

new_data = pd.read_csv("../data/curated/historical_property.csv")
new_data = new_data.drop(columns=['Unnamed: 0'])

school_csv = pd.read_csv("../data/Geo_data/schoolLocations2022.csv",encoding="cp1252")
```

```python
new_data.head()
```

```python
school_csv.head()
```

```python
locations_school = school_csv[['Y','X','School_Name']]

school_location_dict = {}
for i in range(locations_school.shape[0]):
    key = locations_school.iloc[i,2]
    value_Y = locations_school.iloc[i,0]
    value_X = locations_school.iloc[i,1]
    value = [value_Y,value_X]
    school_location_dict[key] = value
school_location_dict
```

```python
target_school_location = school_location_dict['River Nile School']
target_property_location = [new_data.iloc[0,1],new_data.iloc[0,2]]
```

```python
count = 0
small_school_location_list = []

for key,item in school_location_dict.items():
    small_school_location_list.append(item)
    count +=1
    if(count>5):
        break\
            
small_school_location_list

```

# Draw the graph

```python
# Reference from: https://blog.csdn.net/jidawanghao/article/details/124817575

m = folium.Map([-37.840935,144.946457],  zoom_start=10)

for location_school in small_school_location_list:

    ls = folium.PolyLine(locations = [target_property_location,location_school],
                     color='red')
    mark = folium.Marker(location_school)

    ls.add_to(m)
    mark.add_to(m)
    
ls = folium.PolyLine(locations = [target_property_location,target_school_location],
                     color='blue')

mark = folium.Marker(location = target_school_location,icon=folium.Icon(color="green"),)
mark.add_to(m)
mark = folium.Marker(location = target_property_location,icon=folium.Icon(color="red", icon="info-sign"))
mark.add_to(m)

ls.add_to(m)

m.save('../plots/foliumMap_SchoolAPIPresentation_1.html')
m

```

```python
target_school_X = target_school_location[1]
target_school_Y = target_school_location[0]

target_property_location_X = target_property_location[1]
target_property_location_Y = target_property_location[0]
```

```python
(target_school_location,target_property_location)
```

```python
# Reference from: https://medium.com/analytics-vidhya/measure-driving-distance-time-and-plot-routes-between-two-geographical-locations-using-python-39995dfea7e
client = openrouteservice.Client(key='5b3ce3597851110001cf6248cbfec5133fcf4e53a12773fc1404ec0b')

target_school_location = school_location_dict['River Nile School']
target_property_location = [new_data.iloc[0,1],new_data.iloc[0,2]]


coords = ((target_property_location_X,target_property_location_Y),(target_school_X,target_school_Y))
#call API
res = client.directions(coords)
geometry = client.directions(coords)['routes'][0]['geometry']
decoded = convert.decode_polyline(geometry)
```

```python
m_1 = folium.Map([-37.80588,144.96126],  zoom_start=17)

folium.GeoJson(decoded).add_to(m_1)

mark = folium.Marker(location = target_school_location,icon=folium.Icon(color="green"), popup='River Nile School')
mark.add_to(m_1)
mark = folium.Marker(location = target_property_location,icon=folium.Icon(color="red", icon="info-sign"), popup='609/1 BOUVERIE STREET, CARLTON SOUTH')
mark.add_to(m_1)

distance_txt = "<h4> <b>Distance :&nbsp" + "<strong>"+str(round(res['routes'][0]['summary']['distance']/1000,1))+" Km </strong>" +"</h4></b>"
duration_txt = "<h4> <b>Duration :&nbsp" + "<strong>"+str(round(res['routes'][0]['summary']['duration']/60,1))+" Mins. </strong>" +"</h4></b>"

folium.GeoJson(decoded).add_child(folium.Popup(distance_txt+duration_txt,max_width=300)).add_to(m_1)

m_1.save('../plots/foliumMap_SchoolAPIPresentation_2.html')

m_1
```
