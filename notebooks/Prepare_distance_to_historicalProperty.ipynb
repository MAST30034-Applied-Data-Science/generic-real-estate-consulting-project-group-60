# Stored the Property location
The longitude and latitude of each property would be stored at the dictionary

```python
h_property_dict = {}
for i in range(h_property.shape[0]):
    key = h_property.iloc[i,0]
    value = [literal_eval(h_property.iloc[i,1]),literal_eval(h_property.iloc[i,2])]
    h_property_dict[key] = value
```

# Stored the school location
The longitude and latitude of each shool would be stored at the dictionary

```python

school_location_dict = {}
for i in range(locations_school.shape[0]):
    key = locations_school.iloc[i,2]
    value_Y = locations_school.iloc[i,0]
    value_X = locations_school.iloc[i,1]
    value = [value_Y,value_X]
    school_location_dict[key] = value
school_location_dict
```

```python

#Reference from: https://blog.csdn.net/zhuqiuhui/article/details/53180395
# This function would help to get the linear distance by the mathematical function
# @parameters: 
# latA: latitude of position A
# lonA: longitude of position A
# latB: latitude of position B
# lonB: longitude of position B

def getDistance(latA, lonA, latB, lonB):
    ra = 6378140  # radius of equator: meter
    rb = 6356755  # radius of polar: meter
    flatten = (ra - rb) / ra  # Partial rate of the earth
    # change angle to radians
    radLatA = math.radians(latA)
    radLonA = math.radians(lonA)
    radLatB = math.radians(latB)
    radLonB = math.radians(lonB)

    pA = math.atan(rb / ra * math.tan(radLatA))
    pB = math.atan(rb / ra * math.tan(radLatB))
    x = math.acos(math.sin(pA) * math.sin(pB) + math.cos(pA) * math.cos(pB) * math.cos(radLonA - radLonB))
    c1 = (math.sin(x) - x) * (math.sin(pA) + math.sin(pB)) ** 2 / math.cos(x / 2) ** 2
    c2 = (math.sin(x) + x) * (math.sin(pA) - math.sin(pB)) ** 2 / math.sin(x / 2) ** 2
    dr = flatten / 8 * (c1 - c2)
    distance = ra * (x + dr)
    distance = round(distance / 1000, 4)
    return distance


    

     

```

# For the nearest School

```python


# Find the nearest school

for key,item in h_property_dict.items():
    
    distance  = getDistance(item[0], item[1], school_location_dict.get(next(iter(school_location_dict)))[0], school_location_dict.get(next(iter(school_location_dict)))[1])
    school_location_name = next(iter(school_location_dict))
    first_school_Y = school_location_dict.get(next(iter(school_location_dict)))[1]
    first_school_X = school_location_dict.get(next(iter(school_location_dict)))[0]

    
    for skey,sitem in school_location_dict.items():
        new_distance = getDistance(item[0], item[1], sitem[0], sitem[1])
        if (new_distance < distance):
            distance = new_distance
            school_location_name = skey

    
    h_property.iloc[h_property['name']==key,3] = school_location_name
    h_property.iloc[h_property['name']==key,4] = distance





        

```

```python
# Find the nearest station

for key,item in h_property_dict.items():
    
    distance  = getDistance(item[0], item[1], station_point_dict.get(next(iter(station_point_dict)))[0], station_point_dict.get(next(iter(station_point_dict)))[1])
    school_location_name = next(iter(station_point_dict))
    
    first_station_Y = station_point_dict.get(next(iter(station_point_dict)))[1]
    first_station_X = station_point_dict.get(next(iter(station_point_dict)))[0]

    
    for skey,sitem in station_point_dict.items():
        new_distance = getDistance(item[0], item[1], sitem[0], sitem[1])
        if (new_distance < distance):
            distance = new_distance
            station_point_name = skey

    
    h_property.iloc[h_property['name']==key,5] = station_point_name
    h_property.iloc[h_property['name']==key,6] = distance

```
