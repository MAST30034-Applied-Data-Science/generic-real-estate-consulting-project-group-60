Data merge

```python
import pandas as pd
```

```python
data = pd.read_csv("../data/raw/SA2.csv")
data
```

```python
#only for victoria#
data = data.loc[data['S/T code'] == 2]
data
```

```python
#SA2 name and code in VIC#
SA2 = data[['SA2 code', 'SA2 name']]
SA2
```

```python
income = pd.read_csv("../data/raw/2021Census_G32_VIC_SA2.csv")
income
```

```python
income = income[['SA2_CODE_2021','Tot_Tot']]
income
```

```python
income_merge = pd.merge(SA2, income ,how = 'inner', left_on='SA2 code',right_on ='SA2_CODE_2021')
income_merge = income_merge.drop(['SA2_CODE_2021'], axis=1)
income_merge = income_merge.rename(columns={'Tot_Tot':'Total family income weekly'})
income_merge
```

```python
population_2021 = pd.read_csv('../data/raw/population_2021.csv')
population_2021.head()
```

```python
population_2021 = population_2021.loc[population_2021['S/T code'] == 2]
population_2021
```

```python
population_2021 = population_2021[['SA2 code','no..20']]
population_2021 = population_2021.rename(columns={'no..20':'2021 Population'})
population_2021
```

```python
inc_pop_merge = pd.merge(income_merge, population_2021 ,how = 'inner', left_on='SA2 code',right_on ='SA2 code')
#inc_pop_merge = inc_pop_merge.drop(['2021 Population_y'], axis=1)
inc_pop_merge = inc_pop_merge.rename(columns={'2021 Population_x':'2021 Population'})
inc_pop_merge
```

```python
inc_pop_merge=inc_pop_merge.rename(columns={'Total family income weekly':'2021 Income'})
inc_pop_merge['2021 Income']=inc_pop_merge['2021 Income']*52
inc_pop_merge.head()
```
