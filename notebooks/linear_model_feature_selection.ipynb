```python
import pandas as pd
history = pd.read_csv("history_final_merged.csv")
```

```python
history = history.drop(columns=['Unnamed: 0'])
```

```python
current = pd.read_csv("2022_final_merged.csv")
current = current.drop(columns=['Unnamed: 0'])
```

```python
import numpy as np
current['car'] = current['car'].replace(np.nan,0)
history['car'] = history['car'].replace(np.nan,0)
current['bed'] = current['bed'].replace(np.nan,0)
history['bed'] = history['bed'].replace(np.nan,0)
current['bath'] = current['bath'].replace(np.nan,0)
history['bath'] = history['bath'].replace(np.nan,0)
current['property_type'] = current['property_type'].replace(np.nan,'No Mentioned')
history['property_type'] = history['property_type'].replace(np.nan,'No Mentioned')
```

```python
history = history[history['real_school_distance']!='school_distance_txt']
current = current[current['real_distance']!='school_distance_txt']
```

```python
current = current[current['real_distance'].notna()]
history = history[history['real_school_distance'].notna()]
```

```python
COL_FILTER = ['years','property_type','closest_sa2','postcode',  'bed', 'bath','car','Income','Population',
              'week_rent','real_school_distance','real_school_duration','real_station_distance','real_station_duration']
df_filtered = history[COL_FILTER]
```

```python
from statsmodels.formula.api import ols, glm
import statsmodels.api as sm
fit = ols(
    formula="week_rent ~ years +property_type + closest_sa2 +postcode+ bed+bath+car+Income+Population+real_school_distance+real_school_duration+real_station_distance+real_station_duration",
    data=df_filtered,
).fit()
```

```python
print(fit.summary())
```
