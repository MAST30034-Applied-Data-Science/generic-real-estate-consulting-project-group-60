# Housing data 2017-2021

```python
import pandas as pd
import random
from sklearn.model_selection import train_test_split
import numpy as np
from random import randint
#read the data and create the previous housing data
previous_house = pd.read_csv('../data/raw/2022_house_data.csv')
previous_house.head()
```

```python
#change the type of SA2_region back into list form
from ast import literal_eval
previous_house['sa2_region'] = previous_house['sa2_region'].apply(literal_eval)

```

```python
house_17_21 = previous_house[['property_type','coordinates','postcode','sa2_region','bedroom_num','bathroom_num','parking_num']]
#divide these data into 3000 property for each year randomly 3000 for 2017,3000 for 2018, 3000 for 2019, 3000 for 2020, 3000 for 2021
#数据必须要被打乱不然的话每年区域太局限了，保证每年的每个区域每个户型都有一些房子
all_price = []
for i in range(0,15092):
    all_price.append(randint(0,0))
house_17_21['Income'] = all_price
house_17_21['Population'] = all_price
house_17_21['median_price'] = all_price
```

```python
#randomly divide the data and put 3000 into each year 
all_x = house_17_21[['property_type','coordinates','postcode','sa2_region','bedroom_num','bathroom_num','parking_num','Income','Population','median_price']]
all_y = house_17_21['median_price']
X_train, X_test, y_train, y_test = train_test_split(all_x, all_y, test_size=0.8)
house_17 = X_train
all_year = []
for i in range(0,3018):
    all_year.append('2017')
house_17['year_num'] = all_year
house_17
```

```python
X_train, X_test, y_train, y_test = train_test_split(all_x, all_y, test_size=0.8)
house_18 = X_train
all_year = []
for i in range(0,3018):
    all_year.append('2018')
house_18['year_num'] = all_year
house_18
```

```python
X_train, X_test, y_train, y_test = train_test_split(all_x, all_y, test_size=0.8)
house_19 = X_train
all_year = []
for i in range(0,3018):
    all_year.append('2019')
house_19['year_num'] = all_year
house_19
```

```python
X_train, X_test, y_train, y_test = train_test_split(all_x, all_y, test_size=0.8)
house_20 = X_train
all_year = []
for i in range(0,3018):
    all_year.append('2020')
house_20['year_num'] = all_year
house_20
```

```python
X_train, X_test, y_train, y_test = train_test_split(all_x, all_y, test_size=0.8)
house_21 = X_train
all_year = []
for i in range(0,3018):
    all_year.append('2021')
house_21['year_num'] = all_year
house_21
```

```python
frame = [house_17,house_18,house_19,house_20,house_21]
house_17_21_new = pd.concat(frame)
house_17_21_new = house_17_21_new.set_index('year_num')
#now generate the 2017-2021 data
house_17_21_new.to_csv('../data/raw/2017_2021_data.csv')
```
