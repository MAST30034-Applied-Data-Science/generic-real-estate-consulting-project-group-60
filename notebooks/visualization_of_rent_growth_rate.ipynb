```python
# Visualization of rent growth rate
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_history = pd.read_csv('../data/curated/2017_2022_house_data_final.csv')
pred_rent = pd.read_csv('../data/curated/2023_2025_house_data_final.csv')
```

```python
train_history = train_history[train_history.week_rent != 'No Available']
pred_rent = pred_rent[pred_rent.week_rent != 'No Available']
```

```python
train_history = train_history.astype({'week_rent': 'float'})
pred_rent = pred_rent.astype({'week_rent': 'float'})
```

```python
train_history_17 = train_history[train_history.years==2017]
train_history_18 = train_history[train_history.years==2018]
train_history_19 = train_history[train_history.years==2019]
train_history_20 = train_history[train_history.years==2020]
train_history_21 = train_history[train_history.years==2021]
train_history_22 = train_history[train_history.years==2022]
```

```python
pred_23 = pred_rent[pred_rent.year==2023]
pred_24 = pred_rent[pred_rent.year==2024]
pred_25 = pred_rent[pred_rent.year==2025]
```

```python
rent_2017 = train_history_17.groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
rent_2018 = train_history_18.groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
rent_2019 = train_history_19.groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
rent_2020 = train_history_20.groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
rent_2021 = train_history_21.groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
rent_2022 = train_history_22.groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
```

```python
rent_2023 = pred_23.drop(['Unnamed: 0'], axis=1).groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
rent_2024 = pred_24.drop(['Unnamed: 0'], axis=1).groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
rent_2025 = pred_25.drop(['Unnamed: 0'], axis=1).groupby('closest_sa2').mean('week_rent').reset_index()[['closest_sa2','week_rent']]
```

```python
rent_2025 = rent_2025.rename(columns={'week_rent': 'rent_2025'})
rent_2024 = rent_2024.rename(columns={'week_rent': 'rent_2024'})
rent_2023 = rent_2023.rename(columns={'week_rent': 'rent_2023'})
rent_2022 = rent_2022.rename(columns={'week_rent': 'rent_2022'})
rent_2021 = rent_2021.rename(columns={'week_rent': 'rent_2021'})
rent_2020 = rent_2020.rename(columns={'week_rent': 'rent_2020'})
rent_2019 = rent_2019.rename(columns={'week_rent': 'rent_2019'})
rent_2018 = rent_2018.rename(columns={'week_rent': 'rent_2018'})
rent_2017 = rent_2017.rename(columns={'week_rent': 'rent_2017'})
rent_data = [rent_2017,rent_2018,rent_2019,rent_2020,rent_2021,rent_2022,rent_2023]
```

```python
current_rent = rent_2025.merge(rent_2024,on='closest_sa2',how='left')
for each_year in rent_data:
    current_rent = current_rent.merge(each_year,on='closest_sa2',how='left')
current_rent
```

```python
current_rent = current_rent.dropna()
```

```python
current_rent.iloc[1].index
```

```python
import statistics
def get_average_growth(row):
    growth_rate_each_year = []
    last = 'rent_2017'
    curr = 'rent_2017'
    for col in row.index:
        if col == 'closest_sa2':
            continue
        if col == 'rent_2017':
            last = curr
            continue
        curr = col
        growth_rate_each_year.append(row[curr]/row[last])
        last = col
    return statistics.mean(growth_rate_each_year)
```

```python
current_rent.apply(lambda row:get_average_growth(row),axis=1)
```

```python
current_rent['average_growth'] = current_rent.apply(lambda row:get_average_growth(row),axis=1)
```

```python
current_rent.to_csv('../data/curated/average_rent_year.csv')
```
