# 2023-2025缺少邮编为3000的数据

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
from sklearn.metrics import r2_score
```

```python
# this is the final version of the data set
property_data_2017_2022 = pd.read_csv('../data/raw/2017_2022_house_data_final.csv')
property_data_2023_2025 = pd.read_csv('../data/raw/2023_2025_house_data_final.csv')
```

```python
property_data_2023_2025['car'] = property_data_2023_2025['car'].fillna(0)
property_data_2023_2025['bed'] = property_data_2023_2025['bed'].fillna(0)
property_data_2023_2025['bath'] = property_data_2023_2025['bath'].fillna(0)
property_data_2023_2025 = property_data_2023_2025[property_data_2023_2025['real_school_distance'] != 'school_distance_txt']
property_data_2023_2025 = property_data_2023_2025[property_data_2023_2025['real_school_duration'] != 'school_duration_txt']
property_data_2023_2025 = property_data_2023_2025[property_data_2023_2025['real_school_duration'] != 'duration_txt']
property_data_2023_2025 = property_data_2023_2025[property_data_2023_2025['real_station_distance'] != 'station_distance_txt']
property_data_2023_2025 = property_data_2023_2025[property_data_2023_2025['real_station_duration'] != 'station_duration_txt']
property_data_2023_2025 = property_data_2023_2025[property_data_2023_2025['real_station_duration'] != 'duration_txt']

property_data_2017_2022 = property_data_2017_2022[property_data_2017_2022['real_school_distance'] != 'school_distance_txt']
property_data_2017_2022 = property_data_2017_2022[property_data_2017_2022['real_school_duration'] != 'school_duration_txt']
property_data_2017_2022 = property_data_2017_2022[property_data_2017_2022['real_school_duration'] != 'duration_txt']
property_data_2017_2022 = property_data_2017_2022[property_data_2017_2022['real_station_distance'] != 'station_distance_txt']
property_data_2017_2022 = property_data_2017_2022[property_data_2017_2022['real_station_duration'] != 'station_duration_txt']
property_data_2017_2022 = property_data_2017_2022[property_data_2017_2022['real_station_duration'] != 'duration_txt']
```

```python
property_data_2023_2025 = property_data_2023_2025.dropna()
```

```python
X = property_data_2017_2022[['Population','real_station_duration','real_school_duration','real_station_distance','real_school_distance','car','Income','bed','bath','Hospital num']]
y = property_data_2017_2022['week_rent']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
regressor = RandomForestRegressor(n_estimators = 10,random_state = 0)
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
X_future = property_data_2023_2025[['Population','real_station_duration','real_school_duration','real_station_distance','real_school_distance','car','Income','bed','bath','Hospital num']]
```

```python
# predict the future renting fee
y_pred_future = regressor.predict(X_future)
```

```python
property_data_2023_2025['week_rent'] = y_pred_future
```
