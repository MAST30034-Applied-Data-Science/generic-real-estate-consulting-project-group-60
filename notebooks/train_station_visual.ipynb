```python
import folium
import pandas as pd
import geopandas as gpd
```

```python
station_csv = pd.read_csv("../data/raw/datasource-VIC_Govt_PTV-VIC_Govt_DELWP_datavic_PTV_METRO_TRAIN_STATION.csv")
```

```python
station_csv.head()
```

```python
station_point = station_csv.loc[:,["SHAPE","STOP_ID"]]
station_point = station_point.rename(columns={"SHAPE": "geometry","STOP_ID":"LocationID"})
station_point = gpd.GeoDataFrame(station_point)
```

```python
station_point.head()
```

```python
station_point
```

```python
locations = station_csv[['LATITUDE', 'LONGITUDE']]
locationlist = locations.values.tolist()
```

```python
_map = folium.Map(location=[-37.840935,144.946457], zoom_start=10)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=station_point['LocationID'][point]).add_to(_map)
_map
```
